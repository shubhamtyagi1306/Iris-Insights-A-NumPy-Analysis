### Objective: Perform statistical analysis, manipulation, and visualization of a dataset using only Python and the NumPy library.

```python
import numpy as np
```

#### Step 1: Load the data using NumPy script

```python
data = np.genfromtxt('iris.csv', delimiter=',', dtype='str')
data
```

```python
# separate numeric values and labels
features = data[:, :-1].astype(float) # Convert numeric columns to float
labels = data[:, -1] # Extract labels (last column)

print("Features:\n", features[:5])      # Display first 5 rows of numeric data
print("Labels:\n", labels[:5])          # Display first 5 labels
```

### Step 2: Data Cleaning

1. Check for missing values and handle them (e.g., replacing with the mean of the column).
2. Normalize the data for better comparisons.

```python
# Check for missing values
missing = np.isnan(features)

if np.any(missing):
    print('Missing value detected!')

# Replace missing values with the column mean
col_means = np.nanmean(features, axis = 0)
features = np.where(missing, col_means, features)

# Normalize the features (Min-Max scaling)
min_vals = np.min(features, axis=0)
max_vals = np.max(features, axis=0)

normalized_features = (features - min_vals) / (max_vals - min_vals)
print('Normalized Features : \n', normalized_features[:5])
```

### Step 3: Statistical Analysis
1. Calculate basic statistics (mean, median, standard deviation, etc.).
2. Derive insights, such as correlation between features.

```python
# Statistical measures

mean = np.mean(normalized_features, axis=0)
median = np.median(normalized_features, axis=0)
std_devs = np.std(normalized_features, axis=0)

# Correlation matrix
correlation_matrix = np.corrcoef(normalized_features.T)

print('Mean of each features : \n', mean)
print('Correlation Matrix : \n', correlation_matrix)
```

### Step 4: Visualization using ASCII-style visualizations:

```python
def ascii_histogram(data, bins=10):
    hist, bin_edges = np.histogram(data, bins=bins)
    for i, count in enumerate(hist):
        print(f"{bin_edges[i]:.2f} - {bin_edges[i+1]:.2f} | {'#' * count}")

print('Histogram of the first feature : ')
ascii_histogram(normalized_features[:, 0])
```
